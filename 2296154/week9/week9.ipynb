### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

### Step 2. Import the dataset and Assign it to a variable titanic 

```python
titanic = pd.read_csv('titanic.csv')
titanic.head()
```

### Step 3. Set PassengerId as the index 

```python
titanic.set_index("PassengerId")
```

### Step 4. Create a pie chart presenting the male/female proportion

```python
# Grouping the Titanic dataset by 'Sex' and counting the number of passengers in each category
sex_counts = titanic.groupby('Sex').count()

# Plotting a pie chart using the 'PassengerId' column and setting the legend location
sex_counts.plot(y="PassengerId", kind="pie", autopct='%1.1f%%').legend(loc="lower right")
```

### Step 5. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
#Create a relational/scatter plot using Seaborn
sns.relplot(data=titanic, y="Fare", x="Age", hue="Sex")
plt.show()
```

### Step 6. How many people survived?

```python
# Display the count of passengers who survived (1) and did not survive (0)
titanic["Survived"].value_counts()
```

### Step 7. Create a histogram with the Fare payed

```python
# Plotting histogram for the 'Fare' column in the Titanic dataset
titanic["Fare"].plot.hist()
# Adding title and labels to the plot
plt.title("Distribution of Fare in Titanic Dataset")

# Display the plot
plt.show()
```
